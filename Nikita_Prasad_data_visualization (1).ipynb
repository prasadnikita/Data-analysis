```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data= pd.read_csv("Customer-Churn.csv")
data.head()
```

SCATTERPLOT

```python
x=data['tenure'].dropna().head(50)
y=data['TotalCharges'].dropna().head(50)
plt.scatter(x,y,color='indigo')
plt.xlabel("Tenure of the Customer_Churn")
plt.ylabel("Count")
plt.title("Distribution of Contract")
```

BAR PLOT

```python
x=data['Contract'].value_counts().keys()
y=data['Contract'].value_counts()
plt.bar(x,y,color='orange')
plt.xlabel('Contract type of customers')
plt.ylabel('Count of Customers')
plt.title('Distribution of contract') 
```

BOX PLOT

```python
df=data.boxplot(column='MonthlyCharges',by=['PaymentMethod'],figsize=(10,5),color='olive',showmeans=True)
plt.xlabel('Payment Method of customer')
plt.ylabel("Monthly charges incurred")
plt.title("Monthly charges Vs Payment method")
```

```python
df=data.boxplot(column='MonthlyCharges',by=['PaymentMethod'],grid=False,figsize=(10,5),showmeans=True)
```

```python
plt.hist(data['MonthlyCharges'],color='green',bins=50)
plt.xlabel("Monthly Charges incurred ")
plt.ylabel('Count')
plt.title('Distribution of Monthly charges')
plt.grid(True)
```
